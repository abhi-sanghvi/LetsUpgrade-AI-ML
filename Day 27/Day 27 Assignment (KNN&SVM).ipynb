## IMPORTING REQUIRED LIBRARIES

```python
import pandas as pd
from sklearn import preprocessing
from sklearn import neighbors
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
```

## LOADING THE DATASET

```python
data = pd.read_csv('train.csv')
data.head()
```

## DATA CLEANING

```python
# REMOVING THE UNREQUIRED DATA
data1 = data.drop(['PassengerId','Name','Ticket','Cabin'],axis=1)
```

```python
data1.head()
```

## ENCODING THE DATA

```python
lr=preprocessing.LabelEncoder()
data1['Sex']=lr.fit_transform(data1['Sex'])
data1['Embarked']=lr.fit_transform(data1['Embarked'])
```

```python
data1.head()
```

# KNN (K-Nearest Neighbours)

## DEFINING THE FUNCTION

```python
score_ = []
```

```python
def kn(k):
    y = data1.Pclass
    x = data1.drop(['Pclass'],axis=1)
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
    knn = neighbors.KNeighborsClassifier(n_neighbors=k)
    print(knn.fit(x_train,y_train).score(x_test,y_test))
    score_.append(knn.fit(x_train,y_train).score(x_test,y_test))
    y_pred = knn.predict(x_test)
    print(confusion_matrix(y_test,y_pred))
```

```python
a = range(1,268)
```

```python
for i in a:
    print('For K-value =',i)
    kn(i)
    print()
```

```python
plt.plot(a,score_)
```

# SVM (Support Vector Machine)

```python
from sklearn import svm
```

```python
data1.head()
```

## DEFINING THE FUNCTION

```python
def sv(col):
    y=data1[col]
    x=data1.drop([col],axis=1)
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=2)
    clf=svm.SVC(gamma=0.01,C=100)
    y_pred = clf.fit(x_train,y_train).predict(x_test)
    print('The Accuracy score is :-',accuracy_score(y_test,y_pred,normalize=True))
    print('-------------------------------------')
    print('The Confusion matrix is :- \n',confusion_matrix(y_test,y_pred))
```

```python
a = ['Survived','Pclass','Sex','SibSp','Parch','Embarked']
```

```python
for i in a:
    print(i)
    print('================================================')
    sv(i)
    print()
    print('>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>')
```
